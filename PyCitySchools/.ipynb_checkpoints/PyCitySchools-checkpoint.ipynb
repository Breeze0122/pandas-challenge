### District Summary

Calculate:
-Total number of unique schools

-Total students

-Total budget

-Average math score

-Average reading score

-% passing math (the percentage of students who passed math)

-% passing reading (the percentage of students who passed reading)

-% overall passing (the percentage of students who passed math AND reading



```python
import pandas as pd
import os
from functools import reduce

#path of the files prefer use the os library
school_df_path= os.path.join('.','Resources','schools_complete.csv')
student_df_path=os.path.join('.','Resources','students_complete.csv')

#read files
school_df=pd.read_csv(school_df_path, encoding='UTF-8')
student_df=pd.read_csv(student_df_path,encoding='UTF-8')

#Merge the data in a single df

school_df= pd.merge(student_df,school_df, how='left', on=['school_name','school_name'])
school_df.head()

school_df.to_excel('school_df.xlsx',index=False)
```

```python
#Calculate the total number of unique schools
school_count= school_df['school_name'].unique()
school_count= len(school_count)
school_count
```

```python
#Calculate the total number of students
student_count= sum(school_df['student_name'].value_counts())
student_count
```

```python
#Calculate the total budget
total_budget=sum(school_df['budget'].unique())
total_budget
```

```python
#Calculate the average (mean) math score
average_math_score = school_df['math_score'].mean()
average_math_score
```

```python
#Calculate the average (mean) reading score
average_reading_score=school_df['reading_score'].mean()
average_reading_score
    
```

```python
#Calculate % of student that pass math
passing_math_count=school_df[(school_df['math_score']>=70)].count()['student_name']
passing_math_rate= passing_math_count/float(student_count)*100
passing_math_rate
```

```python
#Calculate % of student that pass reading
passing_reading_count=school_df[(school_df['reading_score']>=70)].count()['student_name']
passing_reading_rate = passing_reading_count/float(student_count)*100
passing_reading_rate
```

```python
#calculate % of students that passed math and reading
passing_math_reading_count = school_df[(school_df['math_score']>=70)&(school_df['reading_score']>=70)].count()['student_name']
passing_math_reading_count
overall_passing_rate= passing_math_reading_count/float(student_count)*100
overall_passing_rate
```

```python
#Create a high-level snapshot of district's key metrics in a DataFrame:
district_summ={ 'Total Schools':[school_count],
                'Total Students':[student_count],
                'Total Budget': [total_budget],
                'Avg Math Score': [average_math_score],
                'Avg Reading Score': [average_reading_score],
                '% Passing Math': [passing_math_rate],
                '% Passing Reading': [passing_reading_rate],
                '% Overall passing':[overall_passing_rate]}
district_summ= pd.DataFrame(district_summ,columns=['Total Schools',
                'Total Students',
                'Total Budget',
                'Avg Math Score',
                'Avg Reading Score',
                '% Passing Math',
                '% Passing Reading',
                '% Overall passing'])
# Give format to df
selected_number=['Total Schools',
                'Total Students',
                'Total Budget']

selected_floating=[
                'Avg Math Score',
                'Avg Reading Score']

selected_percentage=[
                '% Passing Math',
                '% Passing Reading',
                '% Overall passing']

#define formating function to each selected group
def format_number(cell_value):
        return '{:,}'.format(cell_value)
    
def format_floating(cell_value):
        return '{:,.2f}'.format(cell_value)
    
def format_percentage(cell_value):
        return '{:,.2f}%'.format(cell_value)

#apply the formating function and display the df

district_summ[selected_number]= district_summ[selected_number].applymap(format_number)
district_summ[selected_floating]=district_summ[selected_floating].applymap(format_floating)
district_summ[selected_percentage]=district_summ[selected_percentage].applymap(format_percentage)
district_summ
```

```python
#Calculate the differenta school names
school_name= school_df['school_name'].unique()
school_name
```

```python
#calculate the school types
school_types= school_df.drop_duplicates(subset='school_name')[['school_name','type']]
school_types= school_types.set_index('school_name')
school_types
```

```python
#Calculate the total student count per school
per_school_counts=school_df['school_name'].value_counts().to_frame().rename(columns={'count':'Total Students'})
per_school_counts
```

```python
#Calculate the total school budget and per captia spending
per_school_budget=school_df.groupby(['school_name','budget'])['student_name'].count().to_frame().rename(columns={'student_name':'#students'})
#add index, wo it, the index error is show ind the div calculation
per_school_budget=per_school_budget.reset_index()
#per_school_budget
#calculate the "per capita spending" and add the column in the same df
per_school_budget['per_capita']=per_school_budget['budget'].div(per_school_budget['#students'],axis=0)
per_school_budget

def format_amount(cell_value):
   return '${:,.2f}'.format(cell_value)

columns_to_format=['budget']
per_school_budget[columns_to_format]= per_school_budget[columns_to_format].applymap(format_amount)
per_school_budget
                                                                  
                                                                  
```

```python
#Calculate the average test scores per school
per_school_math=school_df.groupby('school_name')['math_score'].mean().to_frame()
per_school_math
per_school_reading=school_df.groupby('school_name')['reading_score'].mean().to_frame()
per_school_reading
overall_avg_df= per_school_math.merge(per_school_reading,how='inner',on='school_name').rename(columns={'math_score':'avg_math_score','reading_score':'avg_reading_score'})
overall_avg_df

#def format_float (cell_value):
#        return '{:,.2f}'.format(cell_value)
    
#overall_avg_df=overall_avg_df.applymap(format_float)
#overall_avg_df
```

```python
#Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_df[school_df['math_score']>=70]
students_passing_math = students_passing_math.groupby('school_name')['math_score'].count().to_frame().rename(columns={'math_score':'#students'})
students_passing_math
```

```python
#Calculate the number of students per school with reading scrores of 70 or higher
students_passing_reading=school_df[school_df['reading_score']>=70]
students_passing_reading=students_passing_reading.groupby('school_name')['reading_score'].count().to_frame().rename(columns={'reading_score':'#students'})
students_passing_reading
```

```python
#Calculate the provided code to calculate the number of studens that pass math and reading
students_passing_math_and_reading = school_df[(school_df['reading_score']>=70) & (school_df['math_score']>=70)]
students_passing_math_and_reading = students_passing_math_and_reading.groupby('school_name').size().to_frame().rename(columns={0:'#students'})
students_passing_math_and_reading
```

```python
#Calculate the passing rates

per_school_passing_math= students_passing_math.merge(per_school_counts, on='school_name',how= 'left')
per_school_passing_math= (per_school_passing_math['#students'].div(per_school_passing_math['Total Students'],axis=0))*100
per_school_passing_reading= students_passing_reading.merge(per_school_counts, on='school_name',how= 'left')
per_school_passing_reading= (per_school_passing_reading['#students'].div(per_school_passing_reading['Total Students'],axis=0))*100
per_school_passing_reading
overal_passing_rate= students_passing_math_and_reading.merge(per_school_counts, on='school_name',how= 'left')
overal_passing_rate=(overal_passing_rate['#students'].div(overal_passing_rate['Total Students'],axis=0))*100
overal_passing_rate                     

#data frame with this info

percentage_school = pd.DataFrame({'% Passing Math':per_school_passing_math,
                                 '% Passing Reading':per_school_passing_reading,
                                 '% Overall Passing':overal_passing_rate})

#def format_perc(cell_value):
#    return '{:,.2f}%'.format(cell_value)

#percentage_school= percentage_school.applymap(format_perc)
percentage_school
```

```python
#Create a only DataFrame called 'per_school_summary' with this columns
data_frames=[school_types,per_school_counts,per_school_budget,overall_avg_df,percentage_school]
per_school_summary= reduce(lambda left,right: pd.merge(left,right,on=['school_name'],how='outer'),data_frames)
per_school_summary= per_school_summary.drop(columns=['#students']).rename(columns={'type':'School Type',
                                                                                   'budget':'Total School Budget',
                                                                                  'per_capita':'Per Student Budget',
                                                                                  'avg_math_score':'Average Math Score',
                                                                                  'avg_reading_score':'Average Reading Score'})
per_school_summary=per_school_summary.set_index('school_name')
per_school_summary =per_school_summary.sort_values('school_name')
per_school_summary
```

### Math Scores by Grade

```python
#Split the data by grade
ninth_graders = school_df[(school_df["grade"] == "9th")]
tenth_graders = school_df[(school_df["grade"] == "10th")]
eleventh_graders = school_df[(school_df["grade"] == "11th")]
twelfth_graders = school_df[(school_df["grade"] == "12th")]



#Group by 'school_name'and take the mean of the 'math_score' column for each

ninth_grade_math_scores= ninth_graders.groupby('school_name')['math_score'].mean().to_frame().rename(columns={'school_name':'9th'})
tenth_grader_math_scores = tenth_graders.groupby('school_name')['math_score'].mean().to_frame().rename(columns={'school_name':'10th'})
eleventh_grader_math_scores = eleventh_graders.groupby('school_name')['math_score'].mean().to_frame().rename(columns={'school_name':'11th'})
twelfth_grader_math_scores = twelfth_graders.groupby('school_name')['math_score'].mean().to_frame().rename(columns={'school_name':'12th'})
twelfth_grader_math_scores


# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
data_frames=[ninth_grade_math_scores,tenth_grader_math_scores,eleventh_grader_math_scores]

math_scores_by_grade=reduce(lambda left,right:pd.merge(left,right, on=['school_name'],how='outer'),data_frames)
math_scores_by_grade = math_scores_by_grade.rename(columns={'math_score_x':'9th',
                                                            'math_score_y':'10th',
                                                            'math_score':'11th'})
math_scores_by_grade= math_scores_by_grade.merge(twelfth_grader_math_scores,on='school_name',how='inner').rename(columns={'math_score':'12th'})


# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame

math_scores_by_grade
```

### Reading Score by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_df[(school_df["grade"] == "9th")]
tenth_graders = school_df[(school_df["grade"] == "10th")]
eleventh_graders = school_df[(school_df["grade"] == "11th")]
twelfth_graders = school_df[(school_df["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby('school_name')['reading_score'].mean().to_frame().rename(columns={'school_name':'9th'})
tenth_grader_reading_scores = tenth_graders.groupby('school_name')['reading_score'].mean().to_frame().rename(columns={'school_name':'10th'})
eleventh_grader_reading_scores = eleventh_graders.groupby('school_name')['reading_score'].mean().to_frame().rename(columns={'school_name':'11th'})
twelfth_grader_reading_scores = twelfth_graders.groupby('school_name')['reading_score'].mean().to_frame().rename(columns={'school_name':'12th'})

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
data_frames=[ninth_grade_reading_scores,tenth_grader_reading_scores,eleventh_grader_reading_scores]

reading_scores_by_grade=reduce(lambda left,right:pd.merge(left,right, on=['school_name'],how='outer'),data_frames)

reading_scores_by_grade = reading_scores_by_grade.rename(columns={'reading_score_x':'9th',
                                                            'reading_score_y':'10th',
                                                            'reading_score':'11th'})
reading_scores_by_grade= reading_scores_by_grade.merge(twelfth_grader_reading_scores,on='school_name',how='inner').rename(columns={'reading_score':'12th'})
# Minor data wrangling
#reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th"]]
#reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

### Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels_group = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
# Use 'pd.cut' to categorize spending based on the bins.
school_spending_df['Spending Ranges (Per Student)'] = pd.cut(school_spending_df['Per Student Budget'],spending_bins,labels=labels_group,include_lowest=True)
school_spending_df


```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean().to_frame()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean().to_frame()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean().to_frame()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean().to_frame()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean().to_frame()
```

```python
# Assemble into DataFrame
spending_summary = spending_math_scores.merge(spending_reading_scores,on='Spending Ranges (Per Student)',how='inner').merge(spending_passing_math,on='Spending Ranges (Per Student)',how='inner').merge(spending_passing_reading,on='Spending Ranges (Per Student)',how='inner').merge(overall_passing_spending,on='Spending Ranges (Per Student)',how='inner')

# Display results
spending_summary
```

### Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels_student = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary['School Size'] = pd.cut(per_school_summary['Total Students'],size_bins,labels=labels_student,include_lowest=False)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean().to_frame()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean().to_frame()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean().to_frame()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean().to_frame()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean().to_frame()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = size_math_scores.merge(size_reading_scores,on='School Size',how='inner').merge(size_passing_math,on='School Size',how='inner').merge(size_passing_reading,on='School Size',how='inner').merge(size_overall_passing,on='School Size',how='inner')

# Display results
size_summary
```

### Scores by School Type

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = per_school_summary.groupby(["School Type"])["Average Math Score"].mean().to_frame()
average_reading_score_by_type = per_school_summary.groupby(["School Type"])["Average Reading Score"].mean().to_frame()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Type"])["% Passing Math"].mean().to_frame()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Type"])["% Passing Reading"].mean().to_frame()
average_percent_overall_passing_by_type = per_school_summary.groupby(["School Type"])["% Overall Passing"].mean().to_frame()
```

```python
# Assemble the new data by type into a DataFrame called `type_summary`


data_frames=[average_math_score_by_type,average_reading_score_by_type,average_percent_passing_math_by_type,average_percent_passing_reading_by_type,average_percent_overall_passing_by_type]

type_summary = reduce(lambda left,right:pd.merge(left,right,on=['School Type'],how='outer'),data_frames)

# Display results
type_summary
```
